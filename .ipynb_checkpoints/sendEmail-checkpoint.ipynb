```python
import pandas as pd
from string import Template
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
```

# Send User website calculation boiler

```python
df=pd.read_csv("maillist.csv")
emails=df['Email'].tolist()
Useres=df['Username'].tolist()
df['Password']=df['Password'].astype('str')
Passes=df['Password'].tolist()
```

```python
def read_template(filename):
    with open(filename, 'r', encoding='utf-8') as template_file:
        template_file_content = template_file.read()
    return Template(template_file_content)

MY_ADDRESS='thermal.eff.standard@gmail.com'
PASSWORD='BestenThermal'

s = smtplib.SMTP(host='smtp.gmail.com', port=587)
s.starttls()
s.login(MY_ADDRESS, PASSWORD)

message_template = read_template('message.txt')

for U, P, email in zip(Useres,Passes,emails):
    msg = MIMEMultipart()       # create a message

    # add in the actual person name to the message template
    message = message_template.substitute(PERSON_USER=U,PERSON_PASS=P)
    #print(message)
    
    # setup the parameters of the message
    msg['From']=MY_ADDRESS
    msg['To']=email
    msg['Subject']="Boiler efficiency calculation website"

    # add in the message body
    msg.attach(MIMEText(message, 'plain'))

    # send the message via the server set up earlier.
    s.send_message(msg)
    
    del msg
```

# Send Track code of certificate

```python
df=pd.read_csv("factoryList.csv")
df=df.dropna(subset=['PostCode'])
df=df.loc[df['Status']!="Yes",:]
#df.head()
#df=df.head(2)
emails=df['CoorEmail'].tolist()
Useres=df['CoorName'].tolist()
trackCode=df['PostCode'].tolist()
#print(len(df))
df
```

```python
def read_template(filename):
    with open(filename, 'r', encoding='utf-8') as template_file:
        template_file_content = template_file.read()
    return Template(template_file_content)

MY_ADDRESS='thermal.eff.standard@gmail.com'
PASSWORD='BestenThermal'

s = smtplib.SMTP(host='smtp.gmail.com', port=587)
s.starttls()
s.login(MY_ADDRESS, PASSWORD)

message_template = read_template('certificate_sending_message.txt')

for U, T, email in zip(Useres,trackCode,emails):
    msg = MIMEMultipart()       # create a message

    # add in the actual person name to the message template
    message = message_template.substitute(PERSON_NAME=U,SEND_TRACKCODE=T)
    #print(message,"\n")
    
    # setup the parameters of the message
    msg['From']=MY_ADDRESS
    msg['To']=email
    msg['Subject']="จัดส่งเกียรติบัตรโครงการพัฒนาและสาธิตการใช้เกณฑ์มาตรฐาน ฯ"

    # add in the message body
    msg.attach(MIMEText(message, 'plain'))

    # send the message via the server set up earlier.
    s.send_message(msg)
    
    del msg
```
